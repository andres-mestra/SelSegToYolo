```python
import os
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
model = YOLO("/home/andres/interfaz_sam_to_yolo/data/runs/detect/train/weights/best.pt")
```

```python
dir_images = 'test/'

images_routes = os.listdir(dir_images)
print(len(images_routes))
for image_route in images_routes:
    image = cv2.imread(dir_images + '/' + image_route)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (640,640))
    data = model.predict(image, imgsz=640, conf = 0.6)
    anotaciones = data[0].plot()
    #print(anotaciones)
    plt.imshow(anotaciones)
    plt.show()
    #cv2.imshow('image', anotaciones)
    #show image and prediction using matplotlib
    
 
    
```
